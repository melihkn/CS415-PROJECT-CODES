```python
# --- CONFIGURATION ---
REPO_URL = "https://github.com/melihkn/CS415-PROJECT-CODES.git"
BRANCH = "develop"  # <--- Change this to "dev", "feature-x", etc.
PROJECT_PATH = "/content/drive/MyDrive/CS415/Project"

# --- 1. MOUNT DRIVE ---
from google.colab import drive
import os
import sys

# Mount only if not already mounted
if not os.path.exists('/content/drive'):
    drive.mount('/content/drive')

# --- 2. NAVIGATE TO PROJECT FOLDER ---
# Create the directory if it doesn't exist
if not os.path.exists(PROJECT_PATH):
    os.makedirs(PROJECT_PATH)
    print(f"Created folder: {PROJECT_PATH}")

%cd "{PROJECT_PATH}"

# --- 3. CLONE/PULL & SELECT BRANCH ---
repo_name = REPO_URL.split("/")[-1].replace(".git", "")

if not os.path.exists(repo_name):
    # Case A: Repo doesn't exist yet -> Clone specific branch
    print(f"Cloning {BRANCH} branch...")
    !git clone -b "{BRANCH}" "{REPO_URL}"
else:
    # Case B: Repo exists -> Pull latest & switch branch
    print("Repo exists. Updating and switching branches...")
    %cd "{repo_name}"

    # Ensure we have all branches from origin
    !git fetch origin

    # Switch to the requested branch
    !git checkout "{BRANCH}"

    # Pull latest changes for this branch
    !git pull origin "{BRANCH}"

    %cd ..

# --- 4. ADD TO PYTHON PATH ---
# This allows you to do "import my_model" from your repo files
sys.path.append(os.path.join(PROJECT_PATH, repo_name))

print(f"\nSUCCESS: You are now using branch '{BRANCH}' in {PROJECT_PATH}/{repo_name}")
```
